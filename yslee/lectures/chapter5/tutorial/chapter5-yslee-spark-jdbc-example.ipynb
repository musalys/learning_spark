```python
from pyspark.sql import SparkSession

spark = (SparkSession
         .builder
         .config("spark.jars", "/usr/local/spark/jars/mysql-connector-java-8.0.30.jar")
         .appName("mysql-test")
         .getOrCreate())
spark
```

```python
spark.sparkContext.getConf().getAll()
```

```python
jdbcDF = (spark
          .read
          .format("jdbc")
          .option("url", "jdbc:mysql://mysql:3306/dev")
          .option("driver", "com.mysql.cj.jdbc.Driver")
          .option("dbtable", "test")
          .option("user", "learning_spark")
          .option("password", "spark1234")
          .load())
```
